```python
# make qr code with url https://forms.gle/5Tws7cYbwEe1WAyK6

import qrcode

url = "https://forms.gle/5Tws7cYbwEe1WAyK6"

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)

# URL 데이터 추가
qr.add_data(url)
qr.make(fit=True)

# 이미지 생성
img = qr.make_image(fill_color="black", back_color="white")

# 파일로 저장
img.save("qrcode.png")
print("QR 코드가 'qrcode.png' 파일로 저장되었습니다!")
qr.make_image().save("qr.png")
```
